```python
import pandas as pd
import numpy as np
from qgis.core import *
import folium
pd.set_option("display.max_columns", None)
```

```python
# Load our basemap
latitude = 31.12
longitude = 121.36

sh_map = folium.Map(location=[latitude, longitude], zoom_start=9.5, tiles='Stamen Toner')# Using style
from folium.plugins import HeatMap

# load district boundaries.
file = 'sh_green_roads.geojson'
folium.GeoJson(
    file,
    style_function=lambda feature: {
        'fill': False,
        'color': '#84fab0',
        'weight': 3,
        'dashArray': '1, 1'
    }
).add_to(sh_map)

sh_map
```
